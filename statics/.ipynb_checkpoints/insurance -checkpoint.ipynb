```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

# run and show graph in the same shell
# if not written, a new box will be opened
%matplotlib inline
```

```python
# load the dataset 

df = pd.read_csv("insurance.csv")
```

```python
df.head()
```

```python
# missing value 
df.isnull().sum()
```

```python
# check duplicate and remove them 
df = df.drop_duplicates()
```

```python
df.shape
```

```python
x_axis = ['age', 'bmi', 'children', 'expenses']
for x in x_axis:
    fig, axes = plt.subplots(1, 2, figsize=(18, 4))
    
    # Using histplot instead of the deprecated distplot
    sns.histplot(df[x], ax=axes[0], bins=20)  # bins=20 for histogram customization
    
    # Boxplot remains the same
    sns.boxplot(df[x], ax=axes[1], orient='h', showmeans=True, color='pink')
```

### handle categorical columns 

```python
df.head()
```

```python
df.sex.unique()
```

```python
df['sex'] = df.sex.map({'female':0, 'male':1})
```

```python
df.head()
```

```python
df.smoker.unique()
```

```python
df['smoker'] = df.smoker.map({'yes':1, 'no':0})
```

```python
df.head()
```

```python
df.region.unique()
```

```python
df['region'] = df.region.map({'southwest':1,'southeast':2,'northwest':3,'northeast':4})
```

```python
df.head()
```

```python
df.info()
```

### split the dataset in x and y 

```python
df.columns
```

```python
x = df.drop(['expenses'], axis = 1 )
```

```python
x.head()
```

```python
y = df[['expenses']]
```

```python
y.head()
```

### train test split 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

#### Model training

```python
# linear regression 
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
score1 = r2_score(y_test, y_pred)
```

```python
score1
```

### support vector regression 

```python
from sklearn.svm import SVR
```

```python
svm = SVR()
```

```python
svm.fit(x_train, y_train)
```

```python
y_pred = svm.predict(x_test)
from sklearn.metrics import r2_score
score1 = r2_score(y_test, y_pred)
```

```python
score1
```

### random forest regression 

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
rf.fit(x_train, y_train)
```

```python
y_pred3 = rf.predict(x_test)
score3 = r2_score(y_test, y_pred3)
```

```python
score3
```

### final model is random forest regressor

```python
rf.fit(x_train, y_train)
```
